````python
from google_play_scraper import reviews, Sort
import pandas as pd

banks = {
    "Commercial Bank of Ethiopia": "com.combanketh.mobilebanking",
    "Bank of Abyssinia": "com.boa.boaMobileBanking",
    "Dashen Bank": "com.dashen.dashensmart"
}

all_data = []

TARGET = 500  

for bank, app_id in banks.items():
    print(f"Scraping {bank}")

    result, _ = reviews(
        app_id,
        lang="en",
        country="et",
        sort=Sort.NEWEST,
        count=TARGET
    )

    for r in result:
        all_data.append({
            "review": r["content"],
            "rating": r["score"],
            "date": r["at"].strftime("%Y-%m-%d"),
            "bank": bank,
            "source": "Google Play"
        })

df = pd.DataFrame(all_data)

print(df.shape)
df.to_csv("../data/raw/raw_reviews.csv", index=False)
````

````python
import pandas as pd

df = pd.read_csv("../data/raw/raw_reviews.csv")

print("Initial shape:", df.shape)

# remove duplicates
df = df.drop_duplicates()

# drop missing critical fields
df = df.dropna(subset=["review", "rating"])

# normalize date
df["date"] = pd.to_datetime(df["date"]).dt.strftime("%Y-%m-%d")

print("Final shape:", df.shape)

df.to_csv("../data/raw/cleaned_reviews.csv", index=False)

print("Saved cleaned dataset.")
````
